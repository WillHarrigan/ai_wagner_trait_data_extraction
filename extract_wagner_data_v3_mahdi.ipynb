```python
# ----------------------------
# Flag models
# ----------------------------
def FlagField(desc: str):
    return Field(default_factory=Flag, description=desc)

class Flag(BaseModel):
    value: bool = Field(False, description="Set True if the passage clearly supports this flag.")
    description: str = Field("", description="Short justification/quote to justify the flag or 'not mentioned'.")


class GroupFlags(BaseModel):
    # core
    has_core:             Flag = FlagField("Contains core taxonomic/identification fields.")

    # habit
    has_habit:            Flag = FlagField("Passage mentions growth habit/life form (tree/shrub/vine/herb/etc.).")

    # leaves
    has_leaf_morphology:  Flag = FlagField("Describes leaf shape/arrangement/margin/venation morphology.")
    has_leaf_trichomes:   Flag = FlagField("Mentions leaf hair terms (glabrous, tomentose, villous, etc.).")
    has_leaf_measurements:Flag = FlagField("Mentions leaf length/width/petiole measurements (with units).")

    # inflorescence / flowers
    has_inflorescence:    Flag = FlagField("Mentions an inflorescence type (cyme, raceme, panicle, spike, umbel, head, etc.).")
    has_flower_morphology:Flag = FlagField("Describes flower/corolla/calyx/perianth morphology or breeding system.")
    has_flower_measurements: Flag = FlagField("Gives measurements flower components (mm/cm).")

    # specific flower part measurements
    has_corolla_measurements: Flag = FlagField("Gives corolla/petal measurements (length/width/diameter).")
    has_perianth_measurements: Flag = FlagField("Gives perianth measurements (tepals/lobes).")
    has_calyx_measurements:   Flag = FlagField("Gives calyx/sepal measurements.")
    has_bract_measurements:   Flag = FlagField("Gives bract measurements.")
    has_labellum_measurements: Flag = FlagField("Gives labellum/lip measurements (orchids).")

    # sex-specific features
    has_sex_specific_flowers: Flag = FlagField("Mentions staminate/pistillate (male/female) flowers.")
    has_sex_specific_corolla_measurements: Flag = FlagField("Gives separate corolla measurements for male/female flowers.")
    has_sex_specific_perianth_measurements: Flag = FlagField("Gives separate perianth measurements for male/female flowers.")
    has_sex_specific_bract_measurements: Flag = FlagField("Gives separate bract measurements for male/female flowers.")
    has_sex_specific_peduncle_measurements: Flag = FlagField("Gives separate peduncle measurements for male/female flowers.")
    has_sex_specific_male_calyx_measurements: Flag = FlagField("Gives calyx measurements for male flowers.")
    has_sex_specific_female_calyx_measurements: Flag = FlagField("Gives calyx measurements for female flowers.")

    # asteraceae-specific
    has_asteraceae_head:  Flag = FlagField("Capitulum/head with involucre/phyllaries and rays/disks/pappus present.")

    # fruit / seed
    has_fruit:            Flag = FlagField("Mentions fruits or fruit type.")
    has_seed:             Flag = FlagField("Mentions seeds per fruit or seed size measurements.")

    # juvenile
    has_juvenile_leaf:    Flag = FlagField("Mentions juvenile leaves or leaflets (different from adult).")

    # distribution / conservation / cytology
    has_distribution:     Flag = FlagField("Mentions Hawaiian islands/locations or distribution.")
    has_conservation:     Flag = FlagField("Mentions conservation/rarity status.")
    has_cytology:         Flag = FlagField("Mentions ploidy or chromosome numbers.")

    def to_bools(self) -> Dict[str, bool]:
        """Flatten to {flag_name: bool}."""
        return {k: getattr(self, k).value for k in self.model_fields}

    def reasons(self) -> Dict[str, str]:
        """Flatten to {flag_name: description}."""
        return {k: getattr(self, k).description for k in self.model_fields}
```

```python
# ----------------------------
# FLAG_TO_MODEL mapping
# ----------------------------
FLAG_TO_MODEL: Dict[str, tuple] = {
    "has_core":                              ("core",                             CoreFieldsModel),
    "has_habit":                             ("habit",                            HabitFieldsModel),
    "has_leaf_morphology":                   ("leaf_morphology",                  LeafMorphologyFieldsModel),
    "has_leaf_trichomes":                    ("leaf_trichomes",                   LeafTrichomeFieldsModel),
    "has_leaf_measurements":                 ("leaf_measurements",                LeafMeasurementsFieldsModel),
    "has_inflorescence":                     ("inflorescence",                    InflorescenceFieldsModel),
    "has_flower_morphology":                 ("flower_morphology",                FlowerMorphologyFieldsModel),
    "has_flower_measurements":               ("flower_measurements",              FlowerMeasurementsFieldsModel),
    "has_corolla_measurements":              ("corolla_measurements",             CorollaMeasurementsFieldsModel),
    "has_perianth_measurements":             ("perianth_measurements",            PerianthMeasurementsFieldsModel),
    "has_calyx_measurements":                ("calyx_measurements",               CalyxMeasurementsFieldsModel),
    "has_bract_measurements":                ("bract_measurements",               BractMeasurementsFieldsModel),
    "has_labellum_measurements":             ("labellum_measurements",            LabellumMeasurementsFieldsModel),
    "has_sex_specific_flowers":              ("sex_specific_flowers",             SexSpecificFlowerFieldsModel),
    "has_sex_specific_corolla_measurements": ("sex_specific_corolla_measurements",SexSpecificCorollaMeasurementsModel),
    "has_sex_specific_perianth_measurements":("sex_specific_perianth_measurements",SexSpecificPerianthMeasurementsModel),
    "has_sex_specific_bract_measurements":   ("sex_specific_bract_measurements",  SexSpecificBractMeasurementsModel),
    "has_sex_specific_peduncle_measurements":("sex_specific_peduncle_measurements",SexSpecificPeduncleMeasurementsModel),
    "has_sex_specific_male_calyx_measurements":  ("sex_specific_male_calyx_measurements",  SexSpecificMaleCalyxMeasurementsModel),
    "has_sex_specific_female_calyx_measurements":("sex_specific_female_calyx_measurements",SexSpecificFemaleCalyxMeasurementsModel),
    "has_asteraceae_head":                   ("capitulum_head",                   CapitulumHeadFieldsModel),
    "has_fruit":                             ("fruit",                            FruitFieldsModel),
    "has_seed":                              ("seed",                             SeedFieldsModel),
    "has_juvenile_leaf":                     ("juvenile_leaf",                    JuvenileLeafFieldsModel),
    "has_distribution":                      ("distribution",                     DistributionFieldsModel),
    "has_conservation":                      ("distribution",                     DistributionFieldsModel),
    "has_cytology":                          ("cytology",                         CytologyFieldsModel),
}


# ----------------------------
# Generic agent runner
# ----------------------------
async def _run_agent(result_type: type, system_prompt: str, sample: str, model) -> Any:
    agent = Agent(
        model=model,
        output_type=result_type,
        system_prompt=system_prompt,
    )
    resp = await agent.run(sample)
    return resp.output


# ----------------------------
# Orchestrator
# ----------------------------
async def extract_plant(sample: str, model) -> Dict[str, Any]:
    # Phase 1: detect all flags in parallel
    flag_names = list(GroupFlags.model_fields)
    flag_results = await asyncio.gather(*[
        _run_agent(
            Flag,
            f"Botanical specialist. Does the passage contain: {GroupFlags.model_fields[name].description} "
            "Set value=True only if clearly supported. Quote the relevant text.",
            sample, model,
        )
        for name in flag_names
    ], return_exceptions=True)

    flag_values: Dict[str, Flag] = {}
    for name, result in zip(flag_names, flag_results):
        if isinstance(result, Exception):
            print(f"Flag error [{name}]: {result}")
            flag_values[name] = Flag()
        else:
            flag_values[name] = result
    flags = GroupFlags(**flag_values)

    # Phase 2: extract all flagged groups in parallel (deduplicate shared models)
    seen: set = set()
    active: list[tuple[str, type, str]] = [
        (group_name, model_cls, getattr(flags, flag_name).description)
        for flag_name, (group_name, model_cls) in FLAG_TO_MODEL.items()
        if getattr(flags, flag_name).value and group_name not in seen and not seen.add(group_name)
    ]
    active_groups: Dict[str, type] = {group_name: model_cls for group_name, model_cls, _ in active}

    group_results = await asyncio.gather(*[
        _run_agent(
            model_cls,
            f"Botanical specialist. Extract only {group_name} fields. Return null for unsupported fields."
            + (f"\nFlag justification: {flag_desc}" if flag_desc else ""),
            sample, model,
        )
        for group_name, model_cls, flag_desc in active
    ], return_exceptions=True)

    results = {
        group_name: data
        for (group_name, _, _), data in zip(active, group_results)
        if not isinstance(data, Exception)
    }

    # Phase 3: validate and self-correct
    results = await validate_and_correct(sample, model, results, active_groups)

    return results
```

```python
from typing import Literal

# Groups unlikely to hallucinate — skip validation to save tokens
SKIP_VALIDATION = {"core", "distribution", "cytology", "habit"}

# ----------------------------
# Validation models
# ----------------------------
class FieldFlag(BaseModel):
    field: str = Field(..., description= "Field name")
    reason: str = Field(..., description= "Reasoning field was flagged (hallucination, unsupported inference, or constraint violation)")

class GroupValidation(BaseModel):
    is_valid: bool = Field(..., description="False if any hallucinations or clearly unsupported values found")
    flagged_fields: List[FieldFlag] = Field(default_factory=list)

# List-based so the LLM doesn't have to return a dict (avoids ValidationError)
class GroupValidationItem(BaseModel):
    group_name: str
    is_valid: bool = Field(default=True, description="False if any hallucinations or clearly unsupported values found")
    flagged_fields: List[FieldFlag] = Field(default_factory=list)

class BatchValidation(BaseModel):
    groups: List[GroupValidationItem]


# ----------------------------
# Extract species-level text (trim family/genus preamble for correction calls)
# ----------------------------
def extract_species_text(text: str) -> str:
    """Return species-level section only; fall back to full text if not found."""
    import re
    match = re.search(r'\nSpecies:.*', text, re.DOTALL | re.IGNORECASE)
    if match:
        return match.group(0).strip()
    cutoff = int(len(text) * 0.4)
    return text[-cutoff:] if cutoff < len(text) else text


# ----------------------------
# Parallel group re-extractor (used by validate_and_correct for corrections)
# ----------------------------
async def parse_all_groups_parallel(
    sample: str,
    model,
    group_models: Dict,
    group_hints: Dict[str, str] = None,
) -> Dict:
    group_hints = group_hints or {}

    async def run_one(name, cls):
        hint = group_hints.get(name, "")
        prompt = f"Botanical specialist. Extract only {name} fields. Return null for unsupported fields."
        if hint:
            prompt += f"\n{hint}"
        return name, await _run_agent(cls, prompt, sample, model)

    pairs = await asyncio.gather(*[run_one(n, c) for n, c in group_models.items()], return_exceptions=True)
    return {n: d for item in pairs if not isinstance(item, Exception) for n, d in [item]}


# ----------------------------
# Batch validate all groups in ONE call
# ----------------------------
async def validate_all_batch(sample: str, model, results: Dict) -> Dict[str, GroupValidation]:
    """Single LLM call validates all groups. Low-risk groups skipped entirely."""
    to_validate = {n: r for n, r in results.items() if n not in SKIP_VALIDATION}
    if not to_validate:
        return {}

    val_prompt = (
        "You are a botanical data quality auditor. "
        "Given source text and extracted data for multiple field groups, "
        "flag any field that:\n"
        "- contradicts the source text (hallucination)\n"
        "- violates a model constraint (e.g. Measurements min > max)\n"
        "- Cannot be logically inferred by an expert annotator\n"
        "For each group, return an entry with its group_name, is_valid, and any flagged_fields. "
        "Set is_valid=True if no issues found."
    )

    groups_json = json.dumps(
        {name: json.loads(r.model_dump_json()) for name, r in to_validate.items()},
        indent=2,
    )

    user_msg = f"SOURCE TEXT:\n{sample}\n\nEXTRACTED DATA:\n{groups_json}"

    agent = Agent(
        model=model,
        output_type=BatchValidation,
        system_prompt=val_prompt,
    )
    resp = await agent.run(user_msg)

    batch: BatchValidation = resp.output

    returned = {
        item.group_name: GroupValidation(
            is_valid=item.is_valid,
            flagged_fields=item.flagged_fields,
        )
        for item in batch.groups
    }

    # Any group the LLM didn't mention is implicitly valid
    return {
        name: returned.get(name, GroupValidation(is_valid=True))
        for name in to_validate
    }


# ----------------------------
# Validate + self-correct loop
# ----------------------------
async def validate_and_correct(
    sample: str,
    model,
    results: Dict,
    active_groups: Dict,
    group_hints: Dict[str, str] = None,
    max_retries: int = 2,
) -> Dict:
    current = dict(results)
    group_hints = group_hints or {}
    correction_text = extract_species_text(sample)

    for attempt in range(max_retries):
        validations = await validate_all_batch(sample, model, current)
        flagged = {n: v for n, v in validations.items() if not v.is_valid}

        if not flagged:
            print(f"All groups valid" + (f" after {attempt} correction(s)." if attempt else "."))
            return current

        print(f"Attempt {attempt + 1}/{max_retries}: correcting {list(flagged.keys())}")
        for name, val in flagged.items():
            for f in val.flagged_fields:
                print(f"  {name}.{f.field}: {f.reason}")

        correction_hints = {}
        for name in flagged:
            issues = "; ".join(f"{f.field}: {f.reason}" for f in flagged[name].flagged_fields)
            base = group_hints.get(name, "")
            correction_hints[name] = f"{base}\nCorrection notes: {issues}".strip()

        flagged_models = {n: active_groups[n] for n in flagged if n in active_groups}
        corrections = await parse_all_groups_parallel(correction_text, model, flagged_models, group_hints=correction_hints)
        current.update(corrections)

    return current
```
